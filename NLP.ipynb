Bag Of Words

```
a=["Bala","mahesh","chilakala"]
a
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
b=CountVectorizer()
b.fit(a)
b.get_feature_names()
```

```
c=b.transform(a)
```

```
c.toarray()
```

```
import pandas as pd
d=pd.DataFrame(c.toarray())
d.columns=b.get_feature_names()
d
```

TFIDF

```
from sklearn.feature_extraction.text import TfidfVectorizer
m=TfidfVectorizer()
m.fit(a)
m
```

```
n=m.fit_transform(a)
n.toarray()
```

```
pd.DataFrame(n.toarray(),columns=m.get_feature_names())
```

Tokenization


```
nltk.download('punkt')
from nltk.tokenize import word_tokenize
b=word_tokenize(a)
b
```

```
len(b)
```

```
from nltk.probability import FreqDist
f=FreqDist()
```

```
for word in a:
  f[word.lower()]+=1
f
```

```
t=f.most_common(15)
t
```

```
from nltk.tokenize import blankline_tokenize
b=blankline_tokenize(a)
b
```

```
len(b)
```

```
from nltk.util import bigrams,trigrams,ngrams
```

```
bi=list(nltk.bigrams(a))
bi
```

```
tri=list(nltk.trigrams(a))
tri
```

```
ng=list(nltk.ngrams(a,8))
ng
```

Stemming

```
from nltk.stem import PorterStemmer
p=PorterStemmer()
p.stem('having')
```

```
p.stem("bala")
```

```
b=['bala','mahesh','chilakala']
for i in b:
  print(p.stem(i))
```

```
from nltk.stem import LancasterStemmer
l=LancasterStemmer()
l.stem('having')
```

```
for i in b:
  print(l.stem(i))
```

Lemmatization

```
nltk.download('wordnet')
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
lem
```

```
lem.lemmatize('bala')
```

```
for i in b:
  print(lem.lemmatize(i))
```
